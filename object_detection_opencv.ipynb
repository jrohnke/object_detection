# Object detection
In this notebook I implement several different object detection methods.

## OpenCV
I utilise functionality present in OpenCV to create a fast object detection network. I am following this blogpost: http://www.pyimagesearch.com/2017/09/11/object-detection-with-deep-learning-and-opencv/

Combining the [MobileNet](https://arxiv.org/abs/1704.04861) architecture with Single Shot Detectors ([SSDs](https://arxiv.org/abs/1512.02325)) results in fast and accurate object detection. The model is taken from [chuanqi305](https://github.com/chuanqi305/MobileNet-SSD) and was pre-trained on the COCO dataset and fine-tuned on PASCAL VOC.

```python
# Import all the packages
import numpy as np
import argparse
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

model = 'models/MobileNetSSD_deploy.caffemodel'
modeltxt = 'models/MobileNetSSD_deploy.prototxt'

image_path = 'images/image1.jpg'
min_confidence = 0.2

classes = ["background", "aeroplane", "bicycle", "bird", "boat",
    "bottle", "bus", "car", "cat", "chair", "cow", "diningtable",
    "dog", "horse", "motorbike", "person", "pottedplant", "sheep",
    "sofa", "train", "tvmonitor"]
colors = np.random.uniform(0, 255, size=(len(classes), 3))
```

```python
# Load the pre-trained network
net = cv2.dnn.readNetFromCaffe(modeltxt, model)
```

```python
# Load the image and convert it to blob as input to network
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
(h, w) = image.shape[:2]
blob = cv2.dnn.blobFromImage(image, 0.007843, (300, 300), 127.5)
```

```python
# Calculate the predictions from the network
net.setInput(blob)
detections = net.forward()
```

```python
# loop over the detections
image_boxes = image.copy()
for i in np.arange(0, detections.shape[2]):
    # extract the confidence (i.e., probability) associated with the prediction
    confidence = detections[0, 0, i, 2]
 
    # filter out weak detections by ensuring the `confidence` is
    # greater than the minimum confidence
    if confidence > min_confidence:
        # extract the index of the class label from the `detections`,
        # then compute the (x, y)-coordinates of the bounding box for
        # the object
        idx = int(detections[0, 0, i, 1])
        box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
        (startX, startY, endX, endY) = box.astype("int")
 
        # display the prediction
        label = "{}: {:.2f}%".format(classes[idx], confidence * 100)
        print("[INFO] {}".format(label))
        cv2.rectangle(image_boxes, (startX, startY), (endX, endY), colors[idx], 2)
        y = startY - 15 if startY - 15 > 15 else startY + 15
        cv2.putText(image_boxes, label, (startX, y), cv2.FONT_HERSHEY_SIMPLEX, 0.9, colors[idx], 2)
        
fig = plt.figure(figsize=(24,12))
plt.imshow(image_boxes)
plt.axis('off')
plt.show
```
